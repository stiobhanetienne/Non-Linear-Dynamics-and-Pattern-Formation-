4.12 Coexisting Chaotic Saddles and Attractors \\ Invariant sets of a 1D Map

First we start with setup to do the cobweb

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# definition of function - 1D dynamics
def g(x):
  if x < -1/3:
    return (3*x+1)/2
  elif -1/3 < x < 1/3:
    return 3*x
  elif  x > 1/3:
    return 3*x-2

#define the straight h(x)=x
def h(x):
  return x

x_val = np.linspace(-2,2,1000)
y_val_g = np.array([g(x) for x in x_val])
y_val_h = np.array([h(x) for x in x_val])

# Set the x-axis ticks
ticks = np.array([-2, -5/3, -4/3, -1, -2/3, -1/3, 0, 1/3, 2/3, 1, 4/3, 5/3, 2])
tick_labels = ['-2', '-5/3', '-4/3', '-1', '-2/3', '-1/3', '0', '1/3', '2/3', '1', '4/3', '5/3', 2]
```

```python
#plot [-2,2] window
plt.plot(x_val, y_val_g)
plt.plot(x_val, y_val_h)
plt.xticks(ticks, tick_labels)
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid(True)
plt.show()
print("\n", 'Show zoomed in window')
#plot the [-1,1] window
plt.plot(x_val, y_val_g)
plt.plot(x_val, y_val_h)
plt.xticks(ticks, tick_labels)
plt.xlim(-1,1)
plt.ylim(-1,1)
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid(True)

plt.show()
```

a) Initial conditions outside [-1,1]

```python
# Function to create cobweb plot for multiple initial conditions
def cobweb_plot(g, x0_list, iterations_list):
    # Create figure and axis
    fig, ax = plt.subplots()

    # Plot the function g(x) and the line y=x
    ax.plot(x_val, y_val_g, label='g(x)')
    ax.plot(x_val, y_val_h, label='h(x) = x')

    # Colors for different cobwebs
    colors = ['r', 'b', 'g', 'm', 'c', 'y']

    for i, (x0, iterations) in enumerate(zip(x0_list, iterations_list)):
        x_n = x0
        color = colors[i % len(colors)]
        for _ in range(iterations):
            y_n = g(x_n)
            # Plot the vertical line to the function
            ax.plot([x_n, x_n], [x_n, y_n], color, lw=1)
            # Plot the horizontal line to y = x
            ax.plot([x_n, y_n], [y_n, y_n], color, lw=1)
            # Update the current point
            x_n = y_n

    if all(-1 <= x0 <= 1 for x0 in x0_list):
      plt.xlim(-1, 1)
      plt.ylim(-1, 1)

    # Add labels, title, legend, and grid
    plt.xlabel('x')
    plt.ylabel('g(x)')
    plt.grid(True)
    plt.title('Cobweb Plot for g(x)')
    plt.legend()
    plt.show()

# Parameters for the cobweb plot
x0_list = [1.1, -1.1]
iterations_list = [3, 6]

# Generate the cobweb plot for multiple initial conditions
cobweb_plot(g, x0_list, iterations_list)
```

We see that the cob-webbing brings us to -infty (infty) if we start with an IC x0<1 (x0>1).

b) Attractor in [-1,0]

```python
x0_list_att = [-0.8,-0.6,-0.2]
iterations_list_att = [20,20,20]
cobweb_plot(g, x0_list_att, iterations_list_att)
```
